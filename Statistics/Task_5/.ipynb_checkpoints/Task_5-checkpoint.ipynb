```python
import pandas as pd
import scipy
from scipy import stats
from scipy import linalg
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as ss
import pingouin as pg
from scipy.stats import levene
from statsmodels.stats.multicomp import (pairwise_tukeyhsd,
                                         MultiComparison)
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
#Открываем файлы:

df = pd.read_csv('/Users/siarheimasliankou/Documents/Work/DA_Practice/Statistics/Task_5/5_task_1.csv')
df2 = pd.read_csv('/Users/siarheimasliankou/Documents/Work/DA_Practice/Statistics/Task_5/5_task_2.csv')
```

```python
#Проверяем, какой формат фотографий блюд является наилучшим, т.е. в какой из групп в среднем покупали больше всего. 
#Всего пользователям показывались три варианта:

#A – прямоугольные 16:9
#B – квадратные
#C – прямоугольные 12:4

#Используем статитстический тест ANOVA (дисперсионный анализ) для сравнения групп:

gr_a = df.query("group == 'A'").events
gr_b = df.query("group == 'B'").events
gr_c = df.query("group == 'C'").events
```

```python
#Проверим, являются ли дисперсии внутри групп примерно одинаковыми (гомогенными) с помощью теста Левена:

stat, p_value = levene(gr_a, gr_b, gr_c)
```

```python
stat
```

```python
#p-value > 0.05, не отклоняем нулевую гипотезу, дисперсии являются одинаковыми:

p_value
```
